```python
import matplotlib.pyplot as plt
import csv

f = open('/content/born.csv', encoding='cp949')
data = csv.reader(f)

bornList = []
year = []

for row in data :
  x_num = 2000
  if '합계출산율' in row[0] :
    for idx in row[1:] :
      bornList.append(float(idx))
      year.append(x_num)
      x_num+=1

plt.figure(dpi = 130)
plt.style.use('bmh')
plt.plot(bornList, 'r', marker = 'o')
plt.title('2000-2016 합계출산율')
plt.show() 
```
