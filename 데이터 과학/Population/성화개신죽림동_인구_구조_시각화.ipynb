한글 깨짐 처리하기

내가 거주하는 지역(동)의 인구 구조 시각화 하기

```python
import csv
import matplotlib.pyplot as plt

mpl.font_manager._rebuild()

#row[0]: 지역명, row[1]: 총인구수, row[2]: 연령구간인구수, row[3]~  인덱스 부터 연령별 인구 수
f = open('age.csv', encoding = 'cp949')
data = csv.reader(f)

result = []

for row in data :
  if '성화.개신.죽림동' in row[0] :
    for i in row[3:]:
      result.append(int(i)) #인구수 데이터 정수로 변환 후 result 변수에 담기



plt.figure(dpi = 200)
plt.title('청주시 서원구 성화.개신.죽림동')
plt.style.use('classic') #격자 무늬 스타일로 지정 classic, dark_background, ggplot(제일 많이 씀)
plt.plot(result,label ='population')
plt.legend()
plt.show()
```
