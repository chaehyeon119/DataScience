데이터 준비하기

```python
#최고 기온 데이터만 리스트에 담아서 출력하기
import csv
f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] != '' : #최고 기온 데이터 값이 존재하면 (결측치 제외)
    result.append(float(row[-1]))

print(result) #최고 기온 리스트 출력
```

데이터 개수 확인

```python
print(len(result)) #결측치 제외된 데이터
```

데이터 시각화 하기

```python
# 최고 기온 값이 8월에 해당되는 값만 추출해서 그래프 시각화
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

result = []                          #최고 기온 데이터 중 8월 데이터만 저장할 리스트 생성

for row in data:
  if row[-1] !='':                   #최고 기온 결측치 제외
    if row[0].split('-')[1] == '08': #row[0] : 날짜 데이터/ 날짜를 잘라서 08월 데이터만 추출
      result.append(float(row[-1]))  # 결과 리스트에 담기


plt.figure(dpi=200)
plt.plot(result, 'deeppink')
plt.show()
```

매년 생일 기준 최고 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

result = []                          #최고 기온 데이터 중 8월 데이터만 저장할 리스트 생성

for row in data:
  if row[-1] !='':                   #최고 기온 결측치 제외
    if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '09':
      #날짜를 잘라서 매년 06월 09일 데이터 추출 
      result.append(float(row[-1]))  # 결과 리스트에 담기


plt.figure(dpi=200)
plt.plot(result, 'hotpink')
plt.show()
```

매년 생일 기준 최고 기온, 최저 기온 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high = [] #최고 기온
low = []  #최저 기온

for row in data : 
  if row[-1] != '' and row[-2] != '' :
    if row[0].split('-')[1] == '11' and row[0].split('-')[2] == '09' :
      high.append(float(row[-1])) #최고 기온
      low.append(float(row[-2]))  #최저 기온

plt.figure(dpi=200)
plt.plot(high, 'k', label='high') #최고 기온
plt.plot(low, 'y', label='low') #최저 기온
plt.legend()
plt.show()
```

날짜 구간을 좁혀서 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high = [] #최고 기온
low = []  #최저 기온

for row in data : 
  if row[-1] != '' and row[-2] != '' :
    if 1997 <= int(row[0].split('-')[0]): #'1983' 문자열로 넣으면 같은 걸 찾을 수는 있는데 큰 값은 못 찾음
    #뒤 문자열을 int 숫자형으로 바꿔야 크기 비교 가능함
      if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '24':
        high.append(float(row[-1]))
        low.append(float(row[-2]))

#데이터 시각화
#<<차트 설정>>
plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

#====위에 세 줄은 - 기호와 한글 나오는 그래프를 쓸 때 적용해 써라====

#<<plot 차트 그리기>>
plt.plot(high, 'r', label='high') #최고 기온
plt.plot(low, 'k', label='low') #최저 기온
plt.legend() #범례 표시
plt.title("내 생일의 기온 변화 그래프")

plt.show()
```

매년 생일 기준 최고 기온, 평균 기온, 최저 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high = [] #최고 기온
avg = []  #평균 기온
low = []  #최저 기온

for row in data : 
  if row[-1] != '' and row[-2] != '' and row[-3] !='': #최고, 최저 평균 결측치 제거
    if 1997 <= int(row[0].split('-')[0]): #'1983' 문자열로 넣으면 같은 걸 찾을 수는 있는데 큰 값은 못 찾음
    #뒤 문자열을 int 숫자형으로 바꿔야 크기 비교 가능함
      if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '24':
        high.append(float(row[-1]))
        low.append(float(row[-2]))
        avg.append(float(row[-3]))


#데이터 시각화
#<<차트 설정>>
plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

#====위에 세 줄은 '-' 기호와 '한글' 나오는 그래프를 쓸 때 적용해 써라====

#<<plot 차트 그리기>>
plt.plot(high, 'r', label='최고 기온') #최고 기온
plt.plot(low, 'k', label='최저 기온') #최저 기온
plt.plot(avg, 'g', label='평균 기온')
plt.legend() #범례 표시
plt.title("내 생일의 기온 변화 그래프")

plt.show()
```
