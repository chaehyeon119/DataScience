여러 결과를 하나의 그래프 안에 담기

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

plt.title('plotting')
plt.plot([10,20,30,40],label='asc') # 증가를 의미하는 asc(ascending) 범례
plt.plot([40, 30, 20, 10], label='desc') # 감소를 의미하는 desc(descending) 범례

plt.legend(loc=5) # 그래프에 범례 속성 추가
#plt.legend() 
#plt.legend(loc=10)
#디폴트 값은 5임 오른쪽 가운데에 두고 싶으면 loc=10. loc=__ 숫자를 통해 조절함
plt.show() # 그래프 보여주기 
```

기본 색상 약어<br>
r -> red <br>
y -> yello<br>
b -> blue<br>
g -> green<br>
k -> black<br>

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

plt.title('color') # 타이틀 속성 추가

plt.plot([10,20,30,40], color='skyblue', label='skyblue') # 하늘색 그래프 #아래와 같은 결과지만 더 명확한 명시법임.
plt.plot([40, 30, 20, 10], 'pink', label='pink') # 분홍색 그래프

plt.legend() # 그래프에 범례 속성 추가
plt.show() # 그래프 보여주기 
```

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

plt.title('color') # 타이틀 속성 추가

plt.plot([10,20,30,40], color='r', label='red') # 기본 색상은 약자로 가능
plt.plot([40, 30, 20, 10], 'g', label='green') # 분홍색 그래프

plt.legend() # 그래프에 범례 속성 추가
plt.show() # 그래프 보여주기
```

linestyle

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

plt.title('linestyle') # 타이틀 속성 추가

plt.plot([10,20,30,40], color='r', linestyle = '--', label='dashed') #빨간색 dashed 그래프
 # linestyle = '--' -> 라인을 점선으로 인식
plt.plot([40, 30, 20, 10], 'g', ls=':', label='green') # 분홍색 그래프

plt.legend() # 그래프에 범례 속성 추가
plt.show() # 그래프 보여주기
```

marker 추가하기

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

plt.title('linestyle') # 타이틀 속성 추가

plt.plot([10,20,30,40], 'r.', label='circle') #빨간색 dashed 그래프
#color에다가 'r.' 식으로 색+.을 함께쓰면 빨간 점이 나옴
 # linestyle = '--' -> 라인을 점선으로 인식
plt.plot([40, 30, 20, 10], 'g^', label='triangle up') # 분홍색 그래프

plt.legend() # 그래프에 범례 속성 추가
plt.show() # 그래프 보여주기
```

Marker<br>
#r.-- : <색상><마커모양><선모양> 코드로 작성 가능

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

plt.title('linestyle') # 타이틀 속성 추가

plt.plot([10,20,30,40], 'r.-', label='circle') #빨간색 dashed 그래프
#color에다가 'r.' 식으로 색+.을 함께쓰면 빨간 점이 나옴
 # linestyle = '--' -> 라인을 점선으로 인식
plt.plot([40, 30, 20, 10], 'g^--', label='triangle up') # 분홍색 그래프

plt.legend() # 그래프에 범례 속성 추가
plt.show() # 그래프 보여주기
```

데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt # 1) 라이브러리 임포트
f = open('/content/seoul.csv', 'r', encoding='cp949')
data=csv.reader(f)
header=next(data) #헤더 추출

result= [ ] #결측치를 제거한 최고 기온 리스트

for row in data:
  if row[-1] != '' : #최고 기온이 ''(빈문자열)이 아니면,
    result.append(float(row[-1]))

result


plt.figure(dpi=200) #dip를 넣으면 해상도가 커짐
plt.plot(result, 'r')
plt.show()
```
