WordNet 사용해서 synset간 의미로적 유사도 계산하기(0-1 사이 실수)<br>
0이면 두 단어 서로 관계 없음<br>
1이면 완전한 유의어

```python
#고양이, 링스: 시라소니 유사도 계산
x=wn.synset('cat.n.01') #고양기
y=wn.synset('lynx.n.01') #시라소니

#x,y 유사도 계산
x.path_similarity(y)

```

index.html 파일에서(불용어 제외) 가장 많이 등장한 단어 원형 찾아보기


```python
# www.networksciencelab.com 사이트에서 하이퍼 링크 정보 모두 가져오기
from bs4 import BeautifulSoup
from urllib.request import urlopen

with urlopen('http://www.networksciencelab.com/') as doc:
  soup=BeautifulSoup(doc)

#하이퍼링크 추출
links=[(link.string, link['href']) for link in soup.find_all('a') if link.has_attr('href')]

#책 제목만 추출
html_text=[i[0] for i in links] #여기까지는 리스트 형태

html_text
```

```python
#None 데이터 제거
#1단계
 #1,2줄에 None 제거한 값만 가져옴
html_text[2:]
```

```python
#None 데이터 제거 
#2단계
#리스트에서 하나의 텍스트로 변환
html_text=''.join(html_text[2:])
html_text
#join을 사용하면서 리스트는 text 형태로 연결함
```

```python
#하나의 텍스트 -> index.html 파일로 저장
html_file = open('index.html','w')
#'w' -> write용
#'index.html'파일로 열겠다 내 머리도 같이 열리겠다. #경로에 있으면 그 파일을 열고   
html_file.write(html_text)
html_file.close()
```

index.html 파일에서(불용어 제외) 가장 많이 등장한 단어 원형 찾아보기

```python
from bs4 import BeautifulSoup
from collections import Counter
from nltk.corpus import stopwords
from nltk import LancasterStemmer

#형태소 분류기를 생성한다.
ls=nltk.LancasterStemmer()

#파일을 읽고 soup을 만든다.
with open('/content/index.html') as infile:
  soup = BeautifulSoup(infile)
```

```python
soup
```

```python
soup.text
```

```python
#1) 텍스트 추출 -> 토큰화
words=nltk.word_tokenize(soup.text)
words
#단어로 다 분류가 되었음
```

```python
#2) 단어를 소문자로 변환
words=[w.lower() for w in words]
#words에 있는 단어들을 하나씩 가져와서 소문자로 만들어라
words
```

```python
import nltk
nltk.download('stopwords')
  
#3) 불용어, 특수기호를 제거하고 단어의 형태소를 추출
#3-1) 불용어 제거 stopwords.words('english')
#3-2) 특수기호 제거 isalnum()

#isalnum() 문자열이 알파벳([a-zA-Z])과 숫자([0-9])로만 구성

words=[ls.stem(w) for w in words if w not in stopwords.words('english') and w.isalnum()]
#w가 불용어가 아니면서 특수기호가 아닌 것들 
```

```python
words
```

```python
#4) 가장 빈번하게 등장하는 단어 10개 추출
#freqs는 빈번함이라는 약어로 많이 사용됨
freqs=Counter(words)
print(freqs.most_common(10))
```
