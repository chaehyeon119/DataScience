최고 기온 데이터 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] != '' : #최고 기온 데이터 값이 존재하면 (결측치 제외)
    result.append(float(row[-1]))

#최고 기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(result, bins=100, color='r') #bins=> 구간 옵션 100까지 
#100으로 구간을 정해줬지만 온도의 범위가 있는 만큼만 표에 표현이 됨 약 -15~40도 사이
plt.show()
```

```python
#최고 기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(result, bins=1000, color='r') #bins=> 구간 옵션 1000까지 
#구간을 넓혔기 때문에 세세하게 그려짐 구간을 좁히면 막대그래프처럼 그려짐
plt.show()
```

1월과 8월의 데이터를 히스토그램으로 시각화 하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

aug = [] #8월의 최고 기온을 담을 aug 리스트 변수 선언
jan = [] #1월의 최고 기온을 담을 jan 리스트 변수 선언
apr = [] #6월의 최고 기온을 담을 jun 리스트 변수 선언
for row in data:
  month= row[0].split('-')[1] #날짜 데이터에서 월 데이터 추출
  
  if row[-1] != '' :
    if month == '08' : #8월 데이터면 aug 변수에 담기
      aug.append(float(row[-1]))
    if month == '01' : #1월 데이터면 jan 변수에 담기
      jan.append(float(row[-1]))
    if month == '04' : #1월 데이터면 jan 변수에 담기
      apr.append(float(row[-1]))

plt.figure(dpi=200)
plt.hist(aug, bins=100, color='hotpink', label='Aug') 
plt.hist(jan, bins=100, color='skyblue', label= 'Jan')
plt.hist(apr, bins=100, color='g', label= 'Apr')
plt.legend()
plt.show()
```

1월과 8월의 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

aug = [] #최고 기온 데이터를 저장할 리스트 생성
jau = []

for row in data:
  month= row[0].split('-')[1] #날짜 데이터에서 월 데이터 추출
  
  if row[-1] != '' :
    if month == '08' :
      aug.append(float(row[-1]))
    if month == '01' :
      jau.append(float(row[-1]))
    

plt.figure(dpi=200)
plt.boxplot([aug, jau])
#리스트 형태로 그리는 것이 좋음 리스트로 안 그리면 하나 위에 다른 것이 쌓이고 쌓이는 형식으로 그려짐
plt.show()
```

1월부터 12월까지 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

month = [[], [], [], [], [], [], [], [], [], [], [], []] #월별 데이터 저장할 month 변수 선언
#month.append([])

for row in data:
  if row[-1] != '' :
    month[int(row[0].split('-')[1])-1].append(float(row[-1]))
# 월과 같은 번호의 인덱스에 월별 데이터 저장 (예: 1월 row[0].split('-')[1] -> month[0] row[0].split('-')[1]-1 , 2월 -> month[1])


plt.figure(dpi=200)
plt.boxplot(month)
#리스트 형태로 그리는 것이 좋음 리스트로 안 그리면 하나 위에 다른 것이 쌓이고 쌓이는 형식으로 그려짐
plt.show()
```

8월 1일부터 31일까지 최고 기온 데이터 상자 그림 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

day = [] #일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31) :
  day.append([]) #day 리스트 내에 31개 리스트 생성[[], [], [], [],...31개]

for row in data:
  if row[-1] != '' :
    if row[0].split('-')[1] == '08' : # 8월 최고 기온 값 저장 day[0] -> 1일
       day[int(row[0].split('-')[2])-1].append(float(row[-1]))
 
plt.figure(figsize=(10,5), dpi=300)
plt.boxplot(day, showfliers=False) #아웃라이터(이상치) 값 생략 showfliers
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

day = [] #일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31) :
  day.append([]) #day 리스트 내에 31개 리스트 생성[[], [], [], [],...31개]

for row in data:
  if row[-1] != '' :
    if row[0].split('-')[1] == '08' : # 8월 최고 기온 값 저장 day[0] -> 1일
       day[int(row[0].split('-')[2])-1].append(float(row[-1]))
 
plt.style.use('ggplot') #ggplot 격자문의 회색 그래프
# 그래프 스타일 지정 (회색 배경의 격자무늬, 2/4 값을 의미하는 선의 색 변경)
plt.figure(figsize=(10,5), dpi=300)
plt.boxplot(day, showfliers=False) #아웃라이터(이상치) 값 생략 showfliers
plt.show()
```
