```python
import matplotlib.pyplot as plt
import csv

f = open('/content/gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

for row in data :
  if '용담.명암.산성동' in row[0] :
    for idx in range(101) :
      m.append(-int(row[idx+3]))
      f.append(int(row[-(idx+1)]))
f.reverse()

plt.figure(dpi=200)

plt.rc('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.style.use('bmh')
plt.title('청주시 용담.명암.산성동의 남녀 성별 인구 분포')
plt.barh(range(101),m, label = "남성")
plt.barh(range(101),f, color = 'r',label = "여성")
plt.legend()

plt.show()
```

#2) 월드컵 베이비 마포구 2002년과 2003년 출생률 비교

```python
import csv
import matplotlib.pyplot as plt

mpl.font_manager._rebuild()
#"시점","행정구역(시군구)별","성별",계

f = open('/content/worldCup2003.csv', encoding = 'cp949')
data = csv.reader(f)

#노원구
nowon = [] #노원구



for row in data :
  if '노원' in row[1] :
    for i in row[4:]:
      nowon.append(float(i.replace(',',''))) #인구수 데이터 정수로 변환 후 nowon 변수에 담기
print(nowon)
#인구구조 시각화
plt.figure(dpi = 200)
plt.style.use('ggplot')
plt.rc('font', family='NanumBarunGothic')

plt.title('노원구 2003년 인구 구조') 
plt.plot(nowon, label = '노원구 인구')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

name = '노원구'

f = open('/content/worldCup2002.csv', encoding= 'cp949')

data2002 =csv.reader(f)

nowon2002 = []

for row in data2002:
  if name in row[1]:
    for i in row[4:] :
      nowon2002.append(float(i.replace(',','')))


f = open('/content/worldCup2003.csv', encoding= 'cp949')

data2003 =csv.reader(f)

nowon2003 = []

for row in data2003:
  if name in row[1]:
    for i in row[4:] :
      nowon2003.append(float(i.replace(',','')))


f = open('/content/worldCup2004.csv', encoding= 'cp949')

data2004 =csv.reader(f)

nowon2004 = []

for row in data2004:
  if name in row[1]:
    for i in row[4:] :
      nowon2004.append(float(i.replace(',','')))

plt.figure(dpi = 130)
plt.style.use('Solarize_Light2') #ggplot 격자 무늬 스타일로 지정# 격자 무늬 스타일로 지정 classic, dark_background

plt.plot(nowon2002,'r', label = '2002')
plt.plot(nowon2003,'b', label = '2003')
plt.plot(nowon2004, 'purple', label = '2004')
plt.title('02-04' + name + '지역의 인구 그래프')
plt.legend()
plt.show() 
```

```python
import matplotlib.pyplot as plt
import csv

f = open('/content/born.csv', encoding='cp949')
data = csv.reader(f)

bornList = []
year = []

for row in data :
  x_num = 2000
  if '합계출산율' in row[0] :
    for idx in row[1:] :
      bornList.append(float(idx))
      year.append(x_num)
      x_num+=1

plt.figure(dpi = 130)
plt.style.use('bmh')
plt.plot(bornList, 'r', marker = 'o')
plt.title('2000-2016 합계출산율')
plt.show() 
```
