제주특별자치도의 성별 분포를 항아리 모양 그래프로 표현

```python
import csv

f = open("gender.csv", encoding="CP949")
data = csv.reader(f)

male = []
female = []

townName = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
for row in data:
    if townName in row[0]:
        for i in row[3:104]: # 남성 데이터 : 3~103번 데이터 슬라이싱 - 정수 데이터로 변환 - male리스트에 저장
            male.append(-int(i.replace(',','')))
        for i in row[106:]: # 남성 데이터 : 106~끝까지 데이터 슬라이싱 - 정수 데이터로 변환 - female리스트에 저장
            female.append(int(i.replace(',','')))
        break

print(male)            
print(female)            

plt.figure(dpi= 200)
plt.title(townName + " 지역의 남녀 성별 인구 분포")
plt.barh(range(101),male, label = "남성")
plt.barh(range(101),female, label = "여성")
plt.legend()
plt.show()
```

제주도의 성별 인구 비율 표현하기

```python
import csv

f = open("gender.csv", encoding="CP949")
data = csv.reader(f)

size = [] #제주도의 남성 인구 합계, 여성 인구 합계

townName = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
for row in data:
    if townName in row[0]:
      m = 0 # 남성 인구수 합계 변수
      f = 0 # 여성 인구수 합계 변수
      for i in range(101) :
        m += int(row[3+i].replace(',', '')) # m = m + int(row[3+i].replace(',', ''))
        f += int(row[106+i].replace(',', ''))
      break

size.append(m)
size.append(f)      


print(size)

#파이차트 시각화
# color = ['crimson', 'darkcyan']
plt.figure(dpi= 200)
label =['남', '여']
plt.axis('equal')
plt.pie(size, labels = label, autopct='%.1f%%',explode=(0,0.1), startangle = 90)
plt.title(townName + '지역의 남녀 성별 비율')
plt.legend()
plt.show()
```
